```python
t20wcs = pd.concat([wc07,wc09,wc10,wc14,wc16,wc21])
```

```python
len(t20wcs)
```

```python
t20wcs.rename({'Group/Semi Final/Final':'Tournament Stage', 'Winner Team':'Winning Team'}, inplace = True, axis = 1)
t20wcs.head()
```

```python
wc12.rename({'Group/Super Eights/Semi Final/Final':'Tournament Stage', 'Winner Team':'Winning Team'},inplace = True, axis =1)
```

```python
wc12.head()
```

```python
t20wcs = pd.concat([t20wcs,wc12])
t20wcs
```

```python
t20wcs['Winning Team Score'].fillna('Null', inplace = True)
t20wcs['Losing Team Score'].fillna('Null', inplace = True)

t20wcs.head()
```

```python
t20wcs['Date'] = pd.to_datetime(t20wcs['Date'])
```

```python
t20wcs['Date']
```

```python
def findyear(row):
    row['Year'] = row['Date'].year
    
    return row

t20wcs = t20wcs.apply(findyear, axis = 1)
```

```python
t20wcs
```

```python
def losing_team(row):
    team1 = row['Match Between'].split(' vs ')[0]
    team2 = row['Match Between'].split(' vs ')[1]
    
    if row['Winning Team'] == team1:
        row['Losing Team'] = team2
    else:
        row['Losing Team'] = team1
        
    return row

t20wcs = t20wcs.apply(losing_team, axis = 1)
    
    
```

```python
def venue_city(row):
    row['City'] = row['Venue'].split(', ')[1]
    return row

t20wcs = t20wcs.apply(venue_city, axis =1)
```

```python
t20wcs.head(50)
```

```python
def wldata(row):
    import re
    col_names = ['Winning Team Score','Losing Team Score']
    for col in col_names:
        if '/' in row[col]:
            score = re.findall('(^[0-9]{1,3})\/([0-9]*)\(([0-9.]*)',row[col])
            if len(score)!= 0:
                row[col.split()[0]+ ' Team Runs'] = score[0][0]
                row[col.split()[0]+ ' Team WIckets'] = score[0][1]
                row[col.split()[0]+ ' Team Overs'] = score[0][2]
        else:
            score = re.findall('(^[0-9]{1,3})\(([0-9.]*)',row[col] )
            if len(score)!= 0:
                row[col.split()[0]+' Team Runs'] = score[0][0]
                row[col.split()[0]+' Team WIckets'] = '10'
                row[col.split()[0]+' Team Overs'] = score[0][1]
            
    return row
            
t20wcs = t20wcs.apply(wldata, axis =1)
```

```python
t20wcs.head(50)
```

```python
margin_of_victory = t20wcs['Result'].str.extract('^.* by ([0-9]*) ([a-z]*)')
margin_of_victory.head()
```

```python
t20wcs['Margin of Victory'] = margin_of_victory[0]
```

```python
t20wcs['Type of Margin'] = margin_of_victory[1]
```

```python
t20wcs.to_csv('All_t20wcs_matches.csv', index=False)
```
